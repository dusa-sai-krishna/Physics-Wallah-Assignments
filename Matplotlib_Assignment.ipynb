```python
import numpy as np
import matplotlib.pyplot as plt
% matplotlib.inline
```

Q1


Matplotlib is a popular plotting library in Python used for creating static, animated, and interactive visualizations in various formats. It provides a wide range of functions and classes for generating high-quality plots, graphs, charts, and other types of visual representations.

Here are five plots that can be created using the Pyplot module of Matplotlib:

Line Plot: A line plot is a simple plot where data points are connected by straight line segments. It is commonly used to represent trends, patterns, or changes over time.

Scatter Plot: A scatter plot displays individual data points as markers, with each marker representing a different data sample. Scatter plots are useful for visualizing relationships between two continuous variables.

Bar Plot: A bar plot (or bar chart) represents categorical data with rectangular bars, where the length of each bar corresponds to the value it represents. Bar plots are often used to compare different categories or display frequency counts.

Histogram: A histogram is a graphical representation of the distribution of a continuous variable. It divides the range of values into bins and displays the frequency or count of data points falling into each bin.

Pie Chart: A pie chart is a circular statistical graphic divided into slices to represent categorical data. Each slice's size is proportional to the quantity it represents, making it useful for illustrating proportions or percentages.

Q2

Scatter Plot: A scatter plot displays individual data points as markers, with each marker representing a different data sample. Scatter plots are useful for visualizing relationships between two continuous variables.

```python
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
print(x,y)
```

```python
plt.scatter(x,y,c='r',marker='p')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('X Vs Y')
```

subplot() is used to create sub axes within a canvas. So a canvas can be divided into multiple axes facilitating multiple plots

```python
fig,axes=plt.subplots(nrows=2,ncols=2)
axes[0,0].plot(np.array([0, 1, 2, 3, 4, 5]),np.array([0, 100, 200, 300, 400, 500]))
axes[0,1].plot(np.array([0, 1, 2, 3, 4, 5]),np.array([50, 20, 40, 20, 60, 70]))
axes[1,0].plot(np.array([0, 1, 2, 3, 4, 5]),np.array([10, 20, 30, 40, 50, 60]))
axes[1,1].plot(np.array([0, 1, 2, 3, 4, 5]),np.array([200, 350, 250, 550, 450, 150]))
```

Q4

Bar Plot: A bar plot (or bar chart) represents categorical data with rectangular bars, where the length of each bar corresponds to the value it represents. Bar plots are often used to compare different categories or display frequency counts.

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.bar(company,profit)
plt.xlabel('company')
plt.ylabel('profit')
```

A box plot, also known as a box-and-whisker plot, is a graphical representation of the distribution of a dataset. It provides a concise summary of the key statistical measures and helps visualize the spread, skewness, and outliers within the data.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
plt.boxplot([box1,box2])
```
