Q1

Seaborn is a powerful Python data visualization library built on top of Matplotlib. It provides a high-level interface for creating aesthetically pleasing statistical graphics. Here are five common plots that can be created using Seaborn, along with their uses:

Line Plot:

Uses: Visualize trends and patterns over time or continuous variables.
Example: Show the change in stock prices over a period of time.
Scatter Plot:

Uses: Display the relationship between two continuous variables.
Example: Plot the correlation between temperature and ice cream sales.
Bar Plot:

Uses: Compare categories or groups by representing data as rectangular bars.
Example: Compare sales figures for different products in a retail store.
Histogram:

Uses: Illustrate the distribution of a single continuous variable.
Example: Analyze the distribution of student scores in an exam.
Heatmap:

Uses: Visualize the correlation between variables in a tabular dataset.
Example: Display the correlation matrix of various economic indicators.
These are just a few examples of the plots that Seaborn can generate. Seaborn offers a wide range of additional plot types, such as box plots, violin plots, and pair plots, among others, each with their own specific uses.
